```python
import numpy as np
import matplotlib.pyplot as plt
from cycler import cycler

from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Tahoma']
```

```python
# Figure 1

FILE = False

custom_cycler = (cycler(color=['gray', 'black', 'darkgrey', 'darkgray','black', 'black', 'gray']))
readX = np.loadtxt('DATA/fig1/timeSeriesX_alpha01_eta1.out', delimiter=',')
xvals = np.arange(0,2000,40)

fig, ax = plt.subplots(1,3, figsize=(18, 6))

ax[0].set_prop_cycle(custom_cycler)
ax[0].plot(xvals, np.array(readX))
ax[0].set_ylabel('Opinions, x')
ax[0].set_xlabel('Time, t')
ax[0].set_xticks([0,500, 1000, 1500,2000])
ax[0].set_xticklabels([0,500,1000, 1500,2000])
ticks=np.array([-10,-5,0,5,10])/10
ax[0].set_yticks(ticks)
ax[0].set_yticklabels(ticks)


readX = np.loadtxt('DATA/fig1/timeSeriesX_alpha02_eta1.out', delimiter=',')

ax[1].set_prop_cycle(custom_cycler)
ax[1].plot(xvals, np.array(readX))
ax[1].set_ylabel('Opinions, x')
ax[1].set_xlabel('Time, t')
ax[1].set_xticks([0,500, 1000, 1500,2000])
ax[1].set_xticklabels([0,500,1000, 1500,2000])
ticks=np.array([-10,-5,0,5,10])
ax[1].set_yticks(ticks)
ax[1].set_yticklabels(ticks)

readX = np.loadtxt('DATA/fig1/timeSeriesX_alpha02_eta4.out', delimiter=',')

ax[2].set_prop_cycle(custom_cycler)
ax[2].plot(xvals, np.array(readX))
ax[2].set_ylabel('Opinions, x')
ax[2].set_xlabel('Time, t')
ax[2].set_xticks([0,500, 1000, 1500,2000])
ax[2].set_xticklabels([0,500,1000, 1500,2000])
ticks=np.array([-10,-5,0,5,10])
ax[2].set_yticks(ticks)
ax[2].set_yticklabels(ticks)

plt.show()
```

```python
# Figure 2

dx, dy = 0.02, 0.3
x = np.arange(0.,0.4+dx,dx)
y = np.arange(-1.0,5.0+dy,dy)
x = np.round(x,decimals=2)
y = np.round(y,decimals=2)
x,y = np.meshgrid(x,y)

z1 = np.loadtxt("DATA/fig2/polarization-100runs.out", delimiter=',') #polarization
z2 = np.loadtxt("DATA/fig2/radicalization-100runs.out", delimiter=',') #radicalization

z1_min, z1_max = 0, abs(z1).max()
z2_min, z2_max = 0, abs(z2).max()

fig, axs = plt.subplots(1, 2, figsize=(13, 6))

ax = axs[0]
c = ax.pcolor(x, y, z1, cmap='viridis', vmin=z1_min, vmax=z1_max, shading='auto')

ax.set_title('Polarization')
cbar = fig.colorbar(c, ax=ax)
ax.set_ylabel('Rewire similarity strength')
ax.set_xlabel('Opinion social influence')

cbar.ax.tick_params(width=2.0) 
cbar.set_label('std(x)')

ticksx = [0.0, 0.1, 0.2, 0.3, 0.4]
ax.set_xticks(ticksx)
ax.set_xticklabels(ticksx)

ticksy = [0, 1, 2, 3, 4, 5]
ax.set_yticks(ticksy)
ax.set_yticklabels(ticksy)

ax = axs[1]
c = ax.pcolor(x, y, z2, cmap='viridis', vmin=z2_min, vmax=z2_max, shading='auto')

ax.set_title('Radicalization')
cbar = fig.colorbar(c, ax=ax)
ax.set_ylabel('Rewire similarity strength, \u03B7')
ax.set_xlabel('Opinion social influence, \u03B1')

ax.set_xticks(ticksx)
ax.set_xticklabels(ticksx)

ax.set_yticks(ticksy)
ax.set_yticklabels(ticksy)

cbar.set_label('avg(|x|)')

fig.tight_layout()
plt.show()
```

```python
# Figure 3

dx, dy = 0.02, 0.3
x = np.arange(0.,0.4+dx,dx)
y = np.arange(-1.0,5.0+dy,dy)
x = np.round(x,decimals=2)
y = np.round(y,decimals=2)
x,y = np.meshgrid(x,y)

z3 = np.loadtxt("DATA/fig3/connectedcomponents-100runs.out", delimiter=',')

z3_min, z3_max = 0, abs(z3).max()

fig, axs = plt.subplots(1, 2, figsize=(13, 6))

ax = axs[0]
c = ax.pcolor(x, y, z3, cmap='viridis', vmin=z3_min, vmax=z3_max)
ax.set_title('Connected components')
cbar = fig.colorbar(c, ax=ax)
ax.set_ylabel('Rewire similarity strength, \u03B7')
ax.set_xlabel('Opinion social influence, \u03B1')


ticksx = [0.0, 0.1, 0.2, 0.3, 0.4]
ax.set_xticks(ticksx)
ax.set_xticklabels(ticksx)

ticksy = [0, 1, 2, 3, 4, 5]
ax.set_yticks(ticksy)
ax.set_yticklabels(ticksy)

ax = axs[1]
ax.plot(np.arange(-1,5.0,0.3), np.transpose(np.array(z3))[10], linewidth=2.0, alpha=1.0, marker='o', markersize=16)

ax.set_ylabel('# connected components')
ax.set_xlabel('Rewire similarity strength, \u03B7')

ticksx = np.round(np.arange(-1,5.6,1.2),1)
ax.set_xticks(ticksx)
ax.set_xticklabels(ticksx)

ticksy = [1, 5, 10, 15, 20, 25, 30]
ax.set_yticks(ticksy)
ax.set_yticklabels(ticksy)

plt.grid(True)

fig.tight_layout()
plt.show()
```
